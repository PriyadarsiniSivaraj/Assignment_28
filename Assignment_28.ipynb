```python
from bs4 import BeautifulSoup
import urllib.request
import nltk
```

```python
# Extracting text from the given webpage
response = urllib.request.urlopen('http://php.net/')
html = response.read()
soup = BeautifulSoup(html,"html5lib")
text = soup.get_text(strip=True)

# Printing the text
print (text)
```

```python
# Tokenizing the given text - i.e., splitting the text into many words
from nltk.tokenize import  word_tokenize
words = word_tokenize(text)
print(words)
```

```python
# Counting the frequency of the words using FreqDist() function
freq = nltk.FreqDist(words) 
for key,val in freq.items(): 
    print (str(key) + ':' + str(val))
```

```python
# Plotting the frequency distribution of the words
freq.plot(20,cumulative=False)
```

```python
# We notice that quite a lot of common words are the frequently used words.So we remove all the stop words to get a better
# picture of the text.

# Removing the stopwords and removing punctuation
import string
from nltk.corpus import stopwords
stopwords.words('english')
clean_tokens = words[:] 

for token in words:
    if token in stopwords.words('english') or token in string.punctuation :
        clean_tokens.remove(token)
```

```python
# Finding the frequency distribution
freq = nltk.FreqDist(clean_tokens) 
for key,val in freq.items(): 
    print (str(key) + ':' + str(val))
```

```python
# Plotting the frequency distribution of the words
freq.plot(20,cumulative=False)
```
